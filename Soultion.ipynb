```python
import pandas as pd

df = pd.read_csv('./dataset/vessel_movements_MBCC.csv', low_memory=False)
pd.set_option('display.max_columns', None)
```

```python
df['operating_mode'] = "No Mode"
df.loc[(df['speed'] > 1) | (df['nav_stat'].isin([0, 3, 4, 8, 11, 12])) & (df['maneuvering_zone'] == 'MBCC Maneouvering Zone'), 'operating_mode'] = "Maneuvering"
df.loc[(df['speed'] < 1) & (pd.notna(df['berth'])), 'operating_mode'] = "Alongside"
df.loc[(df['speed'] > 1) | (df['nav_stat'].isin([0, 3, 4, 8, 11, 12])) & (df['maneuvering_zone'] != 'MBCC Maneouvering Zone'), 'operating_mode'] = "Transit"
df.loc[(df['speed'] < 1) & (df['anchorage'] == 'MBCC Anchorage'), 'operating_mode'] = "Anchorage"


# Filter rows where operating_mode is 'Alongside'
df = df[df['operating_mode'] != 'No Mode']
df.head(5)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df2 = df.sort_values(by=['imo', 'operating_mode', 'timestamp'])
df2.head(5)
```

```python
df2['event_duration'] = df2.groupby('imo')['timestamp'].diff().dt.total_seconds()

df2['event_duration'].fillna(0, inplace=True)

df2['event_duration'] = df2.apply(lambda row: 0 if row['event_duration'] > 10000 else row['event_duration'], axis=1)
```

```python
df2.head(5)
```

```python
df3 = df2[df2['event_duration'] != 0]
df3.head(5)
```

```python
df3.shape
```

```python
df3['date_of_build'] = pd.to_datetime(df3['date_of_build'])
df3.loc[:, 'model_year'] = df3['date_of_build'].dt.year
df3.head(5)
```

```python
df3.loc[df3['model_year'] < 2000, "emission_tier"] = 0
df3.loc[(df3['model_year'] >= 2000) & (df3['model_year'] <= 2010), "emission_tier"] = 1
df3.loc[df3['model_year'] > 2010, "emission_tier"] = 2
df3.head(5)
```

```python
df4 = df3[df3['fuel_category'] == 2]
df4.head(5)
```

```python
df4.shape
```

```python
df4['coverage'].value_counts()
```

```python
# Assuming your DataFrame is named df
columns_to_check = ['p', 'speed', 'ael', 'abl', 'sfc_me', 'sfc_ae']

# Use boolean indexing to filter rows with missing values or values equal to 0 in specified columns
df5 = df4[~df4[columns_to_check].isnull().any(axis=1)]

# Display the resulting DataFrame
df5.head(5)

```

```python
df5.shape
```

```python
df5.loc[:, 'max_speed'] = df5['vref'] * 1.066
df5.loc[:, 'load_factor'] = ((df5['speed'] / df5['max_speed']) ** 3 * 100).round()
df5.head(5)
```

```python
# Assuming your DataFrame is named df5\
# import math
df5['activity_hours'] = df5['event_duration'].apply(lambda x: abs(x / 3600))

# Display the resulting DataFrame
df5.head(5)
```

```python
llaf_table = pd.read_csv('./drive-download-20240126T132517Z-001/supplementary_tables/llaf_table.csv')
llaf_table['Load'] = llaf_table['Load'].str.rstrip('%').astype('float')
df5= pd.merge(df5, llaf_table[['Load', 'CO2']], left_on='load_factor', right_on='Load', how='left')
df5.head(5)
```

```python
df_anchorage = df6[df6['operating_mode'] == 'Anchorage'] 
df_Transit = df6[df6['operating_mode'] == 'Transit'] 
df_Alongside = df6[df6['operating_mode'] == 'Alongside'] 
df_Manugvering = df6[df6['operating_mode'] == 'Maneuvering'] 
```

```python
df_anchorage['emission_factor'] = df6.apply(lambda row: (row['sfc_ae'] + row['sfc_ab']) * 0.867 * 3.667, axis=1)
df_Transit['emission_factor'] = df6.apply(lambda row: (row['sfc_me'] + row['sfc_ae'] + row['sfc_ab']) * 0.867 * 3.667, axis=1)
df_Alongside['emission_factor'] = df6.apply(lambda row: (row['sfc_ae'] + row['sfc_ab']) * 0.867 * 3.667, axis=1)
df_Manugvering['emission_factor'] = df6.apply(lambda row: (row['sfc_me'] + row['sfc_ae'] + row['sfc_ab']) * 0.867 * 3.667, axis=1)
```

```python
Anchorage_C02 = df_anchorage[df_anchorage['operating_mode'] == 'Anchorage'][['ael', 'abl','load_factor', 'activity_hours', 'emission_factor']]
Anchorage_C02['calculated_CO2'] = (Anchorage_C02['ael'] * Anchorage_C02['activity_hours'] * Anchorage_C02['emission_factor']) + (Anchorage_C02['abl'] * Anchorage_C02['activity_hours'] * Anchorage_C02['emission_factor'])
total_Anchorage_C02 = Anchorage_C02['calculated_CO2'].sum()
print(total_Anchorage_C02)
```

```python
Alongside_C02 = df_Alongside[df_Alongside['operating_mode'] == 'Alongside'][['ael', 'abl','load_factor', 'activity_hours', 'emission_factor']]
Alongside_C02['calculated_CO2'] = (Alongside_C02['ael'] * Alongside_C02['activity_hours'] * Alongside_C02['emission_factor']) + (Alongside_C02['abl'] * Alongside_C02['activity_hours'] * Alongside_C02['emission_factor'])
total_Alongside_C02 = Alongside_C02['calculated_CO2'].sum()
print(total_Alongside_C02)
```

```python
Transit_C02 = df_Transit[df_Transit['operating_mode'] == 'Transit'][['p','load_factor','ael', 'abl', 'activity_hours', 'emission_factor','CO2']]
# Transit_C02.activity_hours.unique()
Transit_C02['calculated_CO2'] = (Transit_C02['p'] * Transit_C02['load_factor'] * Transit_C02['activity_hours'] * Transit_C02['emission_factor'] * Transit_C02['CO2']) + (Transit_C02['ael'] * Transit_C02['activity_hours'] * Transit_C02['emission_factor']) + (Transit_C02['abl'] * Transit_C02['activity_hours'] * Transit_C02['emission_factor'])
total_Transit_C02 = Transit_C02['calculated_CO2'].sum()
print(total_Transit_C02)
```

```python
Manugvering_C02 = df_Manugvering[['p','load_factor','ael', 'abl', 'activity_hours', 'emission_factor','CO2']]
# Manugvering_C02.activity_hours.unique()
Manugvering_C02['calculated_CO2'] = (Manugvering_C02['p'] * Manugvering_C02['load_factor'] * Manugvering_C02['activity_hours'] * Manugvering_C02['emission_factor'] * Manugvering_C02['CO2']) + (Manugvering_C02['ael'] * Manugvering_C02['activity_hours'] * Manugvering_C02['emission_factor']) + (Manugvering_C02['abl'] * Manugvering_C02['activity_hours'] * Manugvering_C02['emission_factor'])
t_Manugvering_C02 = Manugvering_C02['calculated_CO2'].sum()
print(t_Manugvering_C02)
```
