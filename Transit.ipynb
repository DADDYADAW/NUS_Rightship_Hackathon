```python
# df3 = df3[['imo','timestamp']]

df3['timestamp'] = pd.to_datetime(df3['timestamp'])
```

```python
df4 = df3.sort_values(by=['imo', 'timestamp'])
```

```python
df4['event_duration'] = df4.groupby('imo')['timestamp'].diff().dt.total_seconds()

df4['event_duration'].fillna(0, inplace=True)

df4['event_duration'] = df4.apply(lambda row: 0 if row['event_duration'] > 10000 else row['event_duration'], axis=1)
```
